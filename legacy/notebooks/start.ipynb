## 使用教学

### 常用入口

- 训练 GUI：打开 AutoDL 暴露的 `6006` 端口。
- 训练监控页：打开 AutoDL 暴露的 `6008` 端口。
- 启动脚本：`/root/autostart_lora_gui.sh`
- 更新并重启脚本：`/root/update_lora_gui.sh`
- 仅重启脚本：`/root/restart_lora_gui.sh`

### 推荐训练流程

1. 运行“启动训练 GUI”。
2. 打开 `6006`，选择训练类型，例如 Anima LoRA / Flux LoRA / SDXL LoRA。
3. 填写训练数据目录和输出目录。
4. 如果使用 Anima，建议开启训练预览图，并保留 `sample_at_first = true`，这样会先生成 step 0 基线图。
5. 打开 `6008` 查看训练进度、Loss 曲线、ETA 和预览图。

### 常见问题

- `6006` 打不开：先运行“仅重启 GUI”，等待 10-30 秒。
- `6008` 没有进度：确认 `6006` 里已经开始训练。
- 日志里出现 `mean ar error`：这是数据分桶的宽高比统计，不是训练报错。
